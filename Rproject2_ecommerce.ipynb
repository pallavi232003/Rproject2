```python

numerical_cols = ['Quantity', 'Unit_Price']
categorical_cols = ['Category', 'Sub_Category', 'Payment_Method', 'Order_Status']

# Plot numerical columns: Histogram + KDE
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True, bins=20)
    plt.title(f'{col}: Histogram & KDE')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

# categorical columns: Countplot
for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=col, data=df)
    plt.title(f'{col}: Count Plot')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Filling categorical columns with mode
for col in ['Category', 'Sub_Category', 'Payment_Method', 'Order_Status']:
    mode_val = df[col].mode()[0]
    df[col].fillna(mode_val, inplace=True)

# Filling Purchase_Date with a placeholder
df['Purchase_Date'].fillna('Unknown', inplace=True)

# Fill Quantity with median
quantity_median = df['Quantity'].median()
df['Quantity'].fillna(quantity_median, inplace=True)

# Fill Unit_Price with random sampling
df['Unit_Price'] = df['Unit_Price'].apply(
    lambda x: x if pd.notnull(x) else df['Unit_Price'].dropna().sample(1).values[0]
)
```

# Trend Analysis: Sales Over Time

```python
# Convert Purchase_Date to datetime (if it's not already) in the analysing part else also by the dataset created person
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'], errors='coerce')

# Group by date: Total Sales & Total Quantity
daily_sales = df.groupby('Purchase_Date').agg({
    'Total_Price': 'sum',
    'Quantity': 'sum'
}).reset_index()


plt.figure(figsize=(10,5))
sns.lineplot(x='Purchase_Date', y='Total_Price', data=daily_sales)
plt.title('Daily Total Sales Trend')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.show()


plt.figure(figsize=(10,5))
sns.lineplot(x='Purchase_Date', y='Quantity', data=daily_sales)
plt.title('Daily Quantity Trend')
plt.xlabel('Date')
plt.ylabel('Total Quantity')
plt.show()
```

# Group By Analysis

```python
category_sales = df.groupby('Category')['Total_Price'].sum().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(x='Category', y='Total_Price', data=category_sales)
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.show()
```

# Profit Margin by Payment Method:

```python
payment_profit = df.groupby('Payment_Method')['Profit_Margin'].mean().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(x='Payment_Method', y='Profit_Margin', data=payment_profit)
plt.title('Average Profit Margin by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Average Profit Margin')
plt.show()
```
